```python
import pandas as pd
import os

output_dir = '/content/drive/MyDrive/University of London/Year 3/FINAL PROJECT/Final Project Code/IOB-NST/Stylized-Outputs'
csv_log_path = os.path.join(output_dir, "nst_results_log.csv")

# Load your experiment log
df = pd.read_csv(csv_log_path)

# Create a composite score
df['composite_score'] = (
    0.4 * df['final_content_loss'] +
    0.6 * df['final_style_loss']
)

# lower losses are better
sorted_df = df.sort_values(by='composite_score', ascending=True)

# Show top N results
top_n = 10
top_results = sorted_df.head(top_n)

# Display selected columns
# print(top_results[[
#     'run_id', 'optimizer', 'init_type',
#     'content_weight', 'style_weight', 'tv_weight',
#     'final_content_loss', 'final_style_loss',
#     'composite_score', 'runtime_sec'
# ]])

top_results.head(10)
```

```python
from IPython.display import Image, display

# Display top result images with metadata
for i, row in top_results.iterrows():
    img_path = row['output_path']  # Full path to saved stylized image
    print(f"Run Index: {row['run_index']} | Optimizer: {row['optimizer']} | Init: {row['init_type']}")
    print(f"Content Weight: {row['content_weight']} | Style Weight: {row['style_weight']} | TV Weight: {row['tv_weight']}")
    print(f"Content Loss: {row['final_content_loss']:.2f}, Style Loss: {row['final_style_loss']:.2f}")
    print(f"Composite Score: {row['composite_score']:.2f} | Runtime: {row['runtime_sec']:.2f}s")

    display(Image(filename=img_path))
    print("-" * 100)
```
